```python
# Import libraries
import pandas as pd

# Create a dataset
# Let's bulld the values lists
value_n_list = ["Andrew", "Phillip", "John", "Maria", "Julian", "Diana", "Oscar", "Pedro", "Daniel", "Sonia" ]
print (value_n_list)

value_w_list = [50, 78, 91.5, 45, 58, 60, 76.3, 55.4, 43,72.7]
value_h_list = [1.76,  1.87,   1.69,   1.54,  1.67,   1.78,   1.89,   1.46,   1.77, 1.90]


data_dict = {"name" : value_n_list,
             "weight": value_w_list,
             "height" : value_h_list }


print(data_dict)

df= pd.DataFrame(data_dict)
print(df)
df
```

```python
df ["bmi"] = df ["weight"] / (df ["height"] ** 2)
df
```

```python
df_sorted = df.sort_values(by=['weight'], ascending=True)
df_sorted

```

```python
def clasificar_bmi(bmi):
  if bmi < 18.5:
    return "bajo peso"
  elif bmi >= 18.5 and bmi < 25:
    return "normal"
  else:
    return "sobrepeso"

df['clasificacion'] = df['bmi'].apply(clasificar_bmi)
print(df)
```
